```python
import cv2
```

```python
import urllib.request
import os

# URL of the Haar Cascade XML file for frontal face detection
haar_cascade_url = 'https://raw.githubusercontent.com/opencv/opencv/4.x/data/haarcascades/haarcascade_frontalface_default.xml'
haar_cascade_filename = 'haarcascade_frontalface_default.xml'

# Download the file if it doesn't exist
if not os.path.exists(haar_cascade_filename):
    print(f'Downloading {haar_cascade_filename}...')
    urllib.request.urlretrieve(haar_cascade_url, haar_cascade_filename)
    print('Download complete.')
else:
    print(f'{haar_cascade_filename} already exists.')
```

```python
haar_features = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img=cv2.imread('/content/ms_dhoni (1) (1).jpeg')
```

```python
#scaleFactor - how much image size is reduced at each step

from google.colab.patches import cv2_imshow

faces = haar_features.detectMultiScale(img,scaleFactor=1.3)
for(x,y,w,h) in faces:
  #img - on which we want to draw rectangle
  # (x,y)- top-left corner
  # (x+w, y+h) -bottom-right corner
  #(0,255,0) - Green color border
  #2-border thickness
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
```

```python
haar_features= cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture(0)
# Note: Direct webcam access (cv2.VideoCapture(0)) often does not work in Google Colab
# due to browser security restrictions. You will likely encounter 'Could not access camera' error.
# For video processing, it is recommended to upload a video file and process it.

while True:
  flag, Frame = cap.read()
  if not flag:
    print("Could not access camera or video stream ended.")
    break

  # Convert frame to grayscale for face detection
  gray_frame = cv2.cvtColor(Frame, cv2.COLOR_BGR2GRAY)

  faces = haar_features.detectMultiScale(gray_frame, scaleFactor=1.3)
  for (x,y,w,h) in faces:
    # Draw rectangles on the original color frame
    cv2.rectangle(Frame,(x,y),(x+w,y+h),(0,255,0),2)

  cv2_imshow(Frame)

  key = cv2.waitKey(25) & 0xFF
  if key == 27:
    break

cap.release()
cv2.destroyAllWindows()
```
